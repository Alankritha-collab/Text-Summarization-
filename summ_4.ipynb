```python
import nltk
import string 
import re
from goose3 import Goose 
import spacy
import pytextrank
g = Goose()
%store -r url
print(url)
#url='https://towardsdatascience.com/text-summarization-with-nlp-textrank-vs-seq2seq-vs-bart-474943efeb09'
article = g.extract(url)

nltk.download('stopwords')
nltk.download('punkt')
stopwords = nltk.corpus.stopwords.words('english')
text=article.cleaned_text

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe(factory_name="textrank")
doc = nlp(text)
en_nlp = spacy.load("en_core_web_sm")
en_nlp.add_pipe("textrank", config={ "stopwords": { "word": ["NOUN"] } })
doc = en_nlp(text)
original_sentences = [sentence for sentence in nltk.sent_tokenize(article.cleaned_text)]
en_nlp = spacy.load("en_core_web_sm")
en_nlp.add_pipe("textrank", config={ "stopwords": { "word": ["NOUN"] } })
doc = en_nlp(text)

tr = doc._.textrank
tr.plot_keyphrases()
summary_4=[]
for sent in tr.summary(limit_phrases=10, limit_sentences=8):
    summary_4.append(sent)
summary_4_string=' '.join([str(item) for item in summary_4])
%store summary_4_string
from IPython.core.display import HTML
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio


def visualize(title,original_sentences,summary_4):
  from IPython.core.display import HTML
  text = ''

  display(HTML(f'<h1>Summary_4- {title}</h1>'))
  for sentence in original_sentences:
    if sentence in summary_4_string:
      text += ' ' + str(sentence).replace(sentence, f"<mark>{sentence}</mark>")
    else:
      text += ' ' + sentence
  display(HTML(f""" {text} """))

visualize(article.title,original_sentences,summary_4_string)
gtts= gTTS(text)
gtts.save('1.wav')
sound_file = '1.wav'
Audio(sound_file, autoplay=True)
```
